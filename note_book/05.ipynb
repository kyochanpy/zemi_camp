```python
import pandas as pd
```

```python
all_df = pd.read_csv('/content/drive/MyDrive/nap_df.csv')
```

```python
only_columns_df = all_df.iloc[:, 5:-3]
```

```python
df_one_hot_encoded = pd.get_dummies(only_columns_df)
```

```python
no_list = []
for i in df_one_hot_encoded.columns:
    if i[-1] == "無":
        no_list.append(i)
```

```python
df_one_hot_encoded = df_one_hot_encoded.drop(no_list, axis=1)
```

```python
column_list = []
for i in df_one_hot_encoded.columns:
    column_list.append(i[:-2])
```

```python
column_list
```

```python
for n in range(0, len(column_list)):
        df_one_hot_encoded.rename(columns={df_one_hot_encoded.columns[n]:column_list[n]}, inplace=True)
```

```python
df_one_hot_encoded
```

```python
nap_df_one_hot_encoded = pd.concat([all_df.iloc[:, :4], df_one_hot_encoded, all_df.iloc[:, -3:]], axis=1)
```

```python
nap_df_one_hot_encoded.to_csv("nap_df_one_hot_encoded.csv", index=False)
```
