```python
def get_num_reviews(url):
    html = getPage(url)
    soup = BeautifulSoup(html, "html.parser")
    str_num_reviews = remove_tags(str(soup.find_all("div", class_="star-point-length")))
    num_reviews = re.sub(r'\D', '', str_num_reviews)

    if num_reviews == "":
        num_reviews = 0
    else:
        num_reviews = int(num_reviews)
    return num_reviews
```

```python
def Confirm_existence_and_get_columns_list(url, id):
    html = getPage(url)
    soup = BeautifulSoup(html, "html.parser")
    disable = soup.find_all("span", class_="is-disabled")

    disable_list = []
    for i in disable:
        str_i = str(i)
        disable_ = p.sub("", str_i)
        disable_ = disable_.replace("\n              ", "")
        disable_ = disable_.replace("\n                ", "")
        disable_ = disable_.replace("  ", "")
        disable_list.append(disable_)

    columns_list = []
    columns_list.append(id)
    for i in all_features_names_without_id:
        if i in disable_list:
            columns_list.append("無")
        else:
            columns_list.append("有")

    return columns_list
```

```python
def get_base_dataset():
    site_ids_list = []
    site_names_list = []
    areas_list = []
    stars_list = []
    for i in list(range(1,39)):
        html = getPage(f"https://www.nap-camp.com/hokkaido/list?sortId=21&pageId={i}")
        soup = BeautifulSoup(html, "html.parser")
        campsite_item = soup.find_all("a", class_="campsite-item")
        for j in list(range(0,10)):
            try:
                id = campsite_item[j].get("href").split('/')[-1]
                name = remove_tags(str(campsite_item[j].find("h3", class_="name")))
                area = remove_tags(str(campsite_item[j].find("div", class_="location")))
                area = area.replace("北海道 &gt; ", "")
                try:
                    star = float(remove_tags((str(campsite_item[j].find("span", class_="point")))))
                except ValueError:
                    star = 0
                site_ids_list.append(id)
                site_names_list.append(name)
                areas_list.append(area)
                stars_list.append(star)
            except IndexError:
                break
    
    output_df = pd.DataFrame()
    output_df["id"] = site_ids_list
    output_df["name"] = site_names_list
    output_df["area"] = areas_list
    output_df["star"] = stars_list

    return output_df
```

```python
def get_features_df(input_df):
    input_df_ = input_df.copy()

    Confirm_existence_lists = []
    num_reviews_list = []
    ids = input_df_["id"].values
    for i in ids:
        try:
            url  = f"https://www.nap-camp.com/hokkaido/{i}"
            Confirm_existence_list = Confirm_existence_and_get_columns_list(url, i)
            Confirm_existence_lists.append(Confirm_existence_list)
            num_reviews = get_num_reviews(url)
            num_reviews_list.append(num_reviews)
        except:
            url  = f"https://www.nap-camp.com/hokkaido/{i}"
            Confirm_existence_list = Confirm_existence_and_get_columns_list(url, i)
            Confirm_existence_lists.append(Confirm_existence_list)
            num_reviews = get_num_reviews(url)
            num_reviews_list.append(num_reviews)

    Confirm_existence_df = pd.DataFrame(Confirm_existence_lists)
    num_reviews_df = pd.DataFrame()
    num_reviews_df["num_reviews"] = num_reviews_list

    for n in range(0, len(all_features_names)):
        Confirm_existence_df.rename(columns={n:all_features_names[n]}, inplace=True)

    features_df = pd.concat([Confirm_existence_df, num_reviews_df], axis=1)

    output_df = pd.merge(input_df_, features_df, on="id")

    return output_df
```
