```python
# coding: utf-8

import requests


def getPage(url):
    res = requests.get(url)
    res.encoding = res.apparent_encoding
    res.encoding = 'utf-8'
    html = res.text
    return html
```

```python
def Confirm_existence_and_get_columns_list(url):
    html = getPage(url)
    soup = BeautifulSoup(html, "html.parser")
    disable = soup.find_all("span", class_="is-disabled")

    disable_list = []
    for i in disable:
        str_i = str(i)
        disable_ = p.sub("", str_i)
        disable_ = disable_.replace("\n              ", "")
        disable_ = disable_.replace("\n                ", "")
        disable_ = disable_.replace("  ", "")
        disable_list.append(disable_)

    columns_list = []
    for a in sisetu_list:
        if a in disable_list:
            columns_list.append("無")
        else:
            columns_list.append("有")
    for b in setubi_list:
        if b in disable_list:
            columns_list.append("無")
        else:
            columns_list.append("有")
    for c in service_list:
        if c in disable_list:
            columns_list.append("無")
        else:
            columns_list.append("有")
    for d in activity_list:
        if d in disable_list:
            columns_list.append("無")
        else:
            columns_list.append("有")

    return columns_list
```

```python
def get_site_ids():
    site_ids_list = []
    for i in list(range(1,39)):
        html = getPage(f"https://www.nap-camp.com/hokkaido/list?sortId=21&pageId={i}")
        soup = BeautifulSoup(html, "html.parser")
        campsite_item = soup.find_all("a", class_="campsite-item")
        print(len(campsite_item))
        for j in list(range(0,10)):
            try:
                id = str(campsite_item[j].find("span", class_="point"))[39:-7]
                site_ids_list.append(id)
            except IndexError:
                break
        
    return site_ids_list
```
